```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.interactive(True)
plt.ion()
matplotlib.is_interactive()
```

```python
data = pd.read_csv('Ecommerce_Purchases.csv')
data.head()
```

```python
data = data.loc[:, ['Job', 'Purchase Price']]
data
```

```python
data = data.groupby(['Job']).sum('Purchase Price').sort_values('Purchase Price', ascending=False)
data = data.head()
# Táº¡o láº¡i cá»t 'Job' báº±ng cÃ¡ch Äáº·t láº¡i chá» má»¥c
data = data.reset_index()
```

```python
#plot
colors = ['blue', 'green', 'red', 'purple', 'orange']
data.plot(x='Job', y='Purchase Price', kind='bar', color=colors)
plt.xlabel('Profession')
plt.ylabel('Total Purchase')
plt.title('Biá»u Äá» cá»t')
```

```python
data = pd.read_csv('Ecommerce_Purchases.csv')
data['LOT2'] = data['Lot'].str.extract('([A-Za-z]+)')
data['LOT2'] = data['LOT2'].str.upper()
data
```

```python
#plot
data = data.groupby(['LOT2']).sum('Purchase Price').sort_values('Purchase Price', ascending=False).head()
# data.plot(kind='bar')
data = data.reset_index()
colors = ['blue', 'green', 'red', 'purple', 'orange']
data.plot(x='LOT2', y='Purchase Price', kind='bar', color=colors)
# plt.xlabel('Profession')
# plt.ylabel('AVG Purchase')
# plt.title('Biá»u Äá» cá»t')
```
